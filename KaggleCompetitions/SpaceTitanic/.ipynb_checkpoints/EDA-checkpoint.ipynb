```python
df["Minor"]=df["Age"].apply(lambda x : 1 if x<18 else 0) 
df["Is.Senior"]=df["Age"].apply(lambda x : 1 if x>60 else 0) 
print("percentage of seniors:", sum(df["Is.Senior"])/df.shape[0] * 100)

print("percentage of survived in seniors:", df.loc[(df["Is.Senior"] == 1) & (df["Transported"] == 1)].shape[0]/df.loc[df["Is.Senior"]==1].shape[0]*100)

print("percentage of minors:", sum(df["Minor"])/df.shape[0] * 100)

print("percentage of survived in minors:", df.loc[(df["Minor"] == 1) & (df["Transported"] == 1)].shape[0]/df.loc[df["Minor"]==1].shape[0]*100)
```
